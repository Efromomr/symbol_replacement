```python
#перевод символа в pil-изображение, а затем в массив numpy

def char_to_pixels(text, path='ARIALUNI.TTF', fontsize=50):
    font = ImageFont.truetype(path, fontsize) 
    w, h = 40, 50
    image = Image.new('L', (w, h), 1)  
    draw = ImageDraw.Draw(image)
    draw.text((0, 0), text, font=font) 
    arr = np.array(image)
    return arr
```

```python
#список из первых 1100 символов Юникода

letters = []
for i in range(0,1100):
    letters.append(chr(i))
```

```python
#словарь из пар символ: матрица пикселей его изображения

data = {}
for letter in letters:
  data[letter] = char_to_pixels(letter)
```

```python
# отдельно названия букв и матрицы букв
l_names = np.array(list(data.keys()))
l_values = np.array(list(data.values()))
l_values = l_values.reshape(1100, -1)
```

```python
#кластеризация

kmeans = KMeans(n_clusters=52,random_state=22)
kmeans.fit(l_values)

groups = {}
for letter, cluster in zip(l_names,kmeans.labels_):
    if cluster not in groups.keys():
        groups[cluster] = []
        groups[cluster].append(letter)
    else:
        groups[cluster].append(letter)
```

```python
#кластеры

for i in range(52):
  print(groups[i])
```

```python
#перевод символов строки в номера кластеров

def chars_to_clusters(string):
  cluster_list = []
  for char in string:
    for i in range(len(groups)):
      if char in groups[i]:
        cluster_list.append(i)
        break
  return cluster_list
```

```python
print(chars_to_clusters('АВСКТ')) #на кириллице 
print(chars_to_clusters('ABCKT')) #на латинице
```
